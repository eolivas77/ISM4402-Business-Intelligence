```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')

Location = "datasets/gradedata.csv"
df = pd.read_csv(Location)
df.head()

df.plot.scatter(x='hours', y='grade')
```

I see one obvious pattern which I think the book is trying to make sure we grasp. The more time we put into our work the better it will be, hence the higher grade. Of course there are always those that fall outside of the norm, for instance those  who got grades between 70-80 while having little to no time invested. On the other hand maybe someone who does not grasp the concept of the class is putting over 15 hours in while still managing to get the lowest grade.
